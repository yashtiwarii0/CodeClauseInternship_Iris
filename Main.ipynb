```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
```

```python
df.head()
```

```python
#check for null values
# This will show the number of null values in each column
df.isnull().sum()
```

```python
#check for shape of the data
# This will show the number of rows and columns in the data
df.shape
```

```python
#check for descriptive statistics
# This will show the count, mean, std, min, 25%, 50%, 75%, max for each column
df.describe()
```

```python
#check for data types
# This will show the data type of each column
df.info()
```

```python
#check for correlation
# This will show the correlation between the columns
new_df = df.drop(columns=['species'])
sns.heatmap(new_df.corr(), annot=True, cmap='copper')
```

```python
#check for unique values in the species column
# This will show the unique values in the species column
df.species.value_counts()
```

```python
#plotting graphs between sepal length and sepal width
# This will show the scatter plot between sepal length and sepal width
ax = plt.axes()
ax.scatter(df.sepal_length , df.sepal_width)
ax.set(xlabel='Sepal Length (cm)' , ylabel='Sepal width (cm)' , title='Sepal Length vs Sepal Width')
plt.show()
```

```python
#plotting pairplot between all the columns
# This will show the pairplot between all the columns
sns.pairplot(df, hue='species', diag_kind='kde', palette='husl')
```

```python
#plotting histogram for sepal length
# This will show the histogram for sepal length
sns.histplot(df['sepal_length'], bins=20, kde=True, color='green', alpha=0.5)
plt.title('Sepal Length Distribution')
```

```python
#plotting histogram for sepal width
# This will show the histogram for sepal width
sns.histplot(df['sepal_width'], bins=20, kde=True, color='blue', alpha=0.5)
plt.title('Sepal Width Distribution')
```

```python
#plotting histogram for petal length
# This will show the histogram for petal length
sns.histplot(df['petal_length'], bins=20, kde=True, color='red', alpha=0.5)
plt.title('Petal Length Distribution')
```

```python
#plotting boxplot for sepal length, sepal width, petal length and petal width
# This will show the boxplot for sepal length, sepal width, petal length and petal width
df.boxplot(column=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], figsize=(10, 5), grid=True)
plt.title('Boxplot of Sepal and Petal Dimensions')
```
